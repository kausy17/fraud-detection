```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

print ("All imports successful!")
```

```python
df = pd.read_csv("../data/raw/creditcard.csv")
print(f"Dataset shape: {df.shape}")
print(f"Rows: {df.shape[0]:,}  |  Columns: {df.shape[1]}")
```

```python
class_counts = df["Class"].value_counts()
fraud_pct = (class_counts[1] / len(df)) * 100

print(f"Legitimate transactions: {class_counts[0]:,}")
print(f"Fraudulent transactions:  {class_counts[1]:,}")
print(f"Fraud percentage:         {fraud_pct:.3f}%")

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].bar(["Legitimate", "Fraud"],
            [class_counts[0], class_counts[1]],
            color=["#4A90D9", "#E24B4A"], edgecolor="white", linewidth=0.5)
axes[0].set_title("Transaction class distribution")
axes[0].set_ylabel("Count")
for i, v in enumerate([class_counts[0], class_counts[1]]):
    axes[0].text(i, v + 500, f"{v:,}", ha="center", fontsize=11)

axes[1].pie([class_counts[0], class_counts[1]],
            labels=["Legitimate", "Fraud"],
            autopct="%1.2f%%",
            colors=["#4A90D9", "#E24B4A"],
            startangle=90)
axes[1].set_title("Class split")

plt.tight_layout()
plt.savefig("../reports/class_distribution.png", dpi=150, bbox_inches="tight")
plt.show()
print("Chart saved to reports/")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

fraud     = df[df["Class"] == 1]["Amount"]
legit     = df[df["Class"] == 0]["Amount"]

axes[0].hist(legit,  bins=100, alpha=0.6, color="#4A90D9", label="Legitimate", density=True)
axes[0].hist(fraud,  bins=100, alpha=0.7, color="#E24B4A", label="Fraud",      density=True)
axes[0].set_xlim(0, 500)   # Focus on transactions under £500
axes[0].set_title("Transaction amount distribution (under £500)")
axes[0].set_xlabel("Amount (£)")
axes[0].set_ylabel("Density")
axes[0].legend()

axes[1].boxplot([legit, fraud],
                labels=["Legitimate", "Fraud"],
                patch_artist=True,
                boxprops=dict(facecolor="#4A90D9", alpha=0.6),
                medianprops=dict(color="black", linewidth=2))
axes[1].set_title("Amount distribution — box plot")
axes[1].set_ylabel("Amount (£)")
axes[1].set_ylim(0, 500)

plt.tight_layout()
plt.savefig("../reports/amount_distribution.png", dpi=150, bbox_inches="tight")
plt.show()

print(f"\nFraud transactions — median amount:     £{fraud.median():.2f}")
print(f"Legitimate transactions — median amount: £{legit.median():.2f}")
print(f"Largest fraud transaction:               £{fraud.max():.2f}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

df["Hour"] = (df["Time"] / 3600) % 24

fraud_hours = df[df["Class"] == 1]["Hour"]
legit_hours = df[df["Class"] == 0]["Hour"]

axes[0].hist(legit_hours, bins=24, alpha=0.6, color="#4A90D9", label="Legitimate", density=True)
axes[0].hist(fraud_hours, bins=24, alpha=0.7, color="#E24B4A", label="Fraud",      density=True)
axes[0].set_title("Transactions by hour of day")
axes[0].set_xlabel("Hour")
axes[0].set_ylabel("Density")
axes[0].legend()

hourly = df.groupby("Hour")["Class"].agg(["sum", "count"])
hourly["fraud_rate"] = hourly["sum"] / hourly["count"] * 100
axes[1].plot(hourly.index, hourly["fraud_rate"], color="#E24B4A", linewidth=2)
axes[1].fill_between(hourly.index, hourly["fraud_rate"], alpha=0.2, color="#E24B4A")
axes[1].set_title("Fraud rate by hour of day (%)")
axes[1].set_xlabel("Hour")
axes[1].set_ylabel("Fraud rate (%)")

plt.tight_layout()
plt.savefig("../reports/time_analysis.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
correlations = df.corr()["Class"].drop("Class").sort_values()

plt.figure(figsize=(10, 8))
colors = ["#E24B4A" if x < 0 else "#4A90D9" for x in correlations]
correlations.plot(kind="barh", color=colors)
plt.title("Feature correlation with fraud (Class = 1)")
plt.xlabel("Correlation coefficient")
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")
plt.tight_layout()
plt.savefig("../reports/feature_correlation.png", dpi=150, bbox_inches="tight")
plt.show()

print("Top 5 features most correlated with fraud:")
print(correlations.abs().sort_values(ascending=False).head())
```
